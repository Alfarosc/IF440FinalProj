```python
import numpy as np

def mean_squared_error(x, y):
  """Calculates the mean squared error between two arrays.

  Args:
    x: The first array.
    y: The second array.

  Returns:
    The mean squared error between the two arrays.
  """
  return np.mean((x - y) ** 2)

def fitness(x, y):
  """Calculates the fitness of a solution using the mean squared error.

  Args:
    x: The solution.
    y: The ground truth.

  Returns:
    The fitness of the solution.
  """
  return 1 - mean_squared_error(x, y)

# x = np.array([1, 0, 0, 1])
# y = np.array([1, 1, 1, 0])

# fitness_score = fitness(x, y)

print(fitness_score)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Define the genetic algorithm parameters
population_size = 10
num_generations = 50
mutation_rate = 0.1

# Create a dataset of known data points
x = np.array([1, 1, 1, 0])
y = np.array([1, 0, 1, 0])

# Reshape the x array
x = x.reshape(-1, 1)

# Define the fitness function
def calculate_fitness(chromosome):
    # Train the model
    model = LinearRegression()
    model.fit(x, chromosome)

    # Predict the values of new data points
    x_new = np.array([1, 1, 1, 1]).reshape(-1, 1)  # Reshape x_new to a 2D array
    y_new = model.predict(x_new)

    # Define the actual target values for the new data points
    y_new_actual = np.array([1, 0, 1, 0])

    # Calculate the MSE
    mse = np.mean((y_new - y_new_actual) ** 2)

    # Fitness is inversely proportional to the MSE
    fitness = 1 / (mse + 1e-6)  # Add a small constant to avoid division by zero

    return fitness

# Initialize the population with random chromosomes
population = np.random.randint(0, 2, size=(population_size, len(x)))

# Main loop for the genetic algorithm
for generation in range(num_generations):
    # Calculate the fitness for each chromosome in the population
    fitness_values = np.array([calculate_fitness(chromosome) for chromosome in population])

    # Select parents based on fitness for crossover
    parents = np.random.choice(np.arange(population_size), size=population_size, p=fitness_values / np.sum(fitness_values), replace=True)

    # Create empty offspring population
    offspring = np.empty_like(population)

    # Perform crossover and mutation to create offspring
    for i in range(0, population_size, 2):
        parent1 = population[parents[i]]
        parent2 = population[parents[i+1]]

        # Perform crossover
        crossover_point = np.random.randint(1, len(x))
        offspring[i] = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        offspring[i+1] = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))

        # Perform mutation
        for j in range(len(x)):
            if np.random.uniform() < mutation_rate:
                offspring[i, j] = 1 - offspring[i, j]
                offspring[i+1, j] = 1 - offspring[i+1, j]

    # Replace the old population with the offspring population
    population = offspring

# Find the best chromosome after the last generation
best_chromosome = population[np.argmax(fitness_values)]

# Train the model with the best chromosome
model = LinearRegression()
model.fit(x, best_chromosome)

# Predict the values of new data points using the best model
x_new = np.array([1, 1, 1, 1]).reshape(-1, 1)  # Reshape x_new to a 2D array
y_new = model.predict(x_new)

# Define the actual target values for the new data points
y_new_actual = np.array([1, 0, 1, 0])

# Calculate the MSE
mse = np.mean((y_new - y_new_actual) ** 2)

# Print the MSE
print("Best MSE:", mse)
print("Best Chromosome:", best_chromosome)
```
